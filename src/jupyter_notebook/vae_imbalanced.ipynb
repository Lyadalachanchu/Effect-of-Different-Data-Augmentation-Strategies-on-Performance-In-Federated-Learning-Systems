```python
from src.update import DatasetSplit
from src.utils import get_dataset
import numpy as np
import matplotlib.pyplot as plt

NUM_CLIENTS, NUM_LABELS = 3, 10

class args:
    def __init__(self, dirichlet):
        self.num_channels = 1
        self.iid = 2
        self.num_classes = 10
        self.num_users = NUM_CLIENTS
        self.dataset = 'mnist'
        self.dirichlet = dirichlet

dirichlet_values = [0.1, 0.5, 10.0, 100.0]

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for i, dirichlet_value in enumerate(dirichlet_values):
    args_instance = args(dirichlet_value)
    dataset, _, user_groups = get_dataset(args_instance)
    clients_labels = np.zeros((NUM_CLIENTS, NUM_LABELS))

    idxs_users = np.random.choice(range(NUM_CLIENTS), NUM_CLIENTS, replace=False)
    for idx in idxs_users:
        idxs = user_groups[idx]
        train_dataset = DatasetSplit(dataset, idxs)
        labels = [t[1].item() for t in train_dataset]
        for label in labels:
            clients_labels[idx][label] += 1

    # Plot for the first client (index 0) in each subplot
    labels = np.arange(NUM_LABELS)
    counts = clients_labels[0]
    ax = axes[i]
    ax.bar(labels, counts, color='blue', alpha=0.7, edgecolor='black')
    ax.set_xticks(labels)
    ax.set_xlabel('Labels')
    ax.set_ylabel('Counts')
    ax.set_title(f'Label Distribution for beta = {dirichlet_value}')
    ax.grid(axis='y', linestyle='--', alpha=0.7)


plt.tight_layout()
plt.show()
```

```python
from collections import Counter
from src.impute import impute_cvae_minority, impute_cvae_naive
from src.vae.mnist_vae import ConditionalVae
import torch
import numpy as np
import matplotlib.pyplot as plt
from src.update import DatasetSplit
from src.utils import get_dataset

NUM_CLIENTS, NUM_LABELS = 100, 10
clients_labels = np.zeros((NUM_CLIENTS, NUM_LABELS))
dirichlet = 0.5

class args:
    def __init__(self):
        self.num_channels = 1
        self.iid = 2
        self.num_classes = 10
        self.num_users = NUM_CLIENTS
        self.dataset = 'mnist'
        self.dirichlet = dirichlet

input_k = 1000
dataset, _, user_groups = get_dataset(args())
idxs_users = np.random.choice(range(NUM_CLIENTS), NUM_CLIENTS, replace=False)
fig, axes = plt.subplots(4, 2, figsize=(14, 18))
print(len(idxs_users))
for row, input_k in enumerate([0,100,500,1000]):
    for idx in idxs_users[:1]:
        print(row)
        idxs = user_groups[idx]
        train_dataset = DatasetSplit(dataset, idxs)
        labels = [t[1].item() for t in train_dataset]
        for label in labels:
            clients_labels[idx][label] += 1

        trained_cvae = ConditionalVae(dim_encoding=3)
        checkpoint = torch.load(f'C:\\Users\\LohithSai\\Desktop\\FederatedImputation\\vae_data\\models\\0_cvae_{dirichlet}_cvae.pth')
        trained_cvae.load_state_dict(checkpoint)

        # print(len(train_dataset))
       # Assuming generated_train_dataset is a list of tuples where the second element is the label
        generated_train_dataset = impute_cvae_minority(k=input_k, trained_cvae=trained_cvae, initial_dataset=train_dataset)
       #  generated_train_dataset = impute_cvae_naive(k=3000, trained_cvae=trained_cvae, initial_dataset=train_dataset)
        post_label_counts = [label for _, label in generated_train_dataset]
        post_label_counts_counter = Counter(post_label_counts)

        # Convert Counter to array
        post_label_counts_array = np.zeros(NUM_LABELS)
        for label, count in post_label_counts_counter.items():
            post_label_counts_array[label] += count


        # print(post_label_counts_array)
        # print("++++++++++++++++++")
        # print(clients_labels[idx])

        # Plot first set of data (before)
        labels = np.arange(NUM_LABELS)
        counts = [x/sum(clients_labels[idx]) for x in clients_labels[idx]]
        axes[row, 0].bar(labels, counts, color='blue', edgecolor='black', alpha=0.7)
        axes[row, 0].set_xticks(labels)
        axes[row, 0].set_xlabel("Digit Label")
        axes[row, 0].set_ylabel("Frequency")
        axes[row, 0].set_title(f"Client {idx + 1} Before k = {input_k}")
        axes[row, 0].grid(axis='y', linestyle='--', alpha=0.7)

        # Plot second set of data (after)
        counts = [x/sum(post_label_counts_array) for x in post_label_counts_array]
        axes[row, 1].bar(labels, counts, color='blue', edgecolor='black', alpha=0.7)
        axes[row, 1].set_xticks(labels)
        axes[row, 1].set_xlabel("Digit Label")
        axes[row, 1].set_ylabel("Frequency")
        axes[row, 1].set_title(f"Client {idx + 1} After k = {input_k}")
        axes[row, 1].grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout
plt.tight_layout()
plt.show()
```
